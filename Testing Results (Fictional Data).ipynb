```python
import matplotlib.pyplot as plt
import numpy as np

def create_perplexity_plot():
    # Separate data for each model
    opt_350m_no_watermark = [2.4979, 2.4979, 2.4979, 2.4979, 2.4979]
    opt_350m_watermarked = [21.9385, 6.5314, 7.1287, 2.9857, 12.8766]
    
    gpt2_medium_no_watermark = [5.2674, 5.2674, 5.2674, 5.2674, 5.27]
    gpt2_medium_watermarked = [8.5382, 3.3683, 5.4811, 7.6624, 1.8021]
    
    # Gamma values
    gamma_values = [0.2, 0.4, 0.6, 0.8, 0.5]

    # Create the plot
    plt.figure(figsize=(8, 5))

    # Plot lines for each model and watermarking condition
    plt.plot(gamma_values, opt_350m_no_watermark, marker='o', label='OPT-350M (No Watermark)', color='blue', linestyle='-')
    plt.plot(gamma_values, opt_350m_watermarked, marker='s', label='OPT-350M (Watermarked)', color='navy', linestyle='--')
    
    plt.plot(gamma_values, gpt2_medium_no_watermark, marker='o', label='GPT-2 Medium (No Watermark)', color='red', linestyle='-')
    plt.plot(gamma_values, gpt2_medium_watermarked, marker='s', label='GPT-2 Medium (Watermarked)', color='darkred', linestyle='--')

    plt.xlabel('Gamma () Values')
    plt.ylabel('Perplexity')
    plt.title('Model Performance Comparison: Perplexity across Gamma Values')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.xticks(gamma_values)

    return plt

# Create and display the plot
plot = create_perplexity_plot()
plot.show()

```
