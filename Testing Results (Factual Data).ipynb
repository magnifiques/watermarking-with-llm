```python
import matplotlib.pyplot as plt

# Plot setup
fig, ax = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Filter data by model
models = df['Model'].unique()
colors = ['blue', 'green', 'orange']

for i, model in enumerate(models):
    model_data = df[df['Model'] == model]
    
    ax[i].plot(model_data['Gamma'], model_data['Bleu Score (Human Generated Text)'], label="Human Generated Text", color=colors[0], marker='o')
    ax[i].plot(model_data['Gamma'], model_data["Bleu Score (Green List / Red List)"], label="Red/Green List", color=colors[1], marker='x')
    ax[i].plot(model_data['Gamma'], model_data["Bleu Score (With Clusters)"], label="Clusters", color=colors[2], marker='s')
    
    ax[i].set_title(f"Model: {model}")
    ax[i].set_xlabel('Gamma')
    ax[i].set_ylabel('BLEU Score')
    ax[i].legend()
    ax[i].grid(True)

# Adjust layout
plt.tight_layout()
# plt.suptitle("Comparison of BLEU Scores Across Models", fontsize=16, y=1.02)
plt.show()
```
